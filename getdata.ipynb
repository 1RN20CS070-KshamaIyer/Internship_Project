```python
import psycopg2
conn = psycopg2.connect(
    host="localhost",
    database="mydatabase",
    user="myuser",
    password="mypassword"
)
```

```python
statements = []

for index,row in df.iterrows():
    statement = f"INSERT INTO Stocks(StockName,Tickers) VALUES ('{row.values[1]}','{row.values[0]}');"
    statements.append(statement)
    print(statement)
```

```python
with conn.cursor() as cursor:
    for statement in statements:
        cursor.execute(statement)
conn.commit()

```
